##Do paid apps have better ratings?

```python
df['Type'].unique()
```

```python
df['Type'].dropna(inplace=True)
```

```python
paid_vs_free_rating=df.groupby('Type')['Rating'].mean()
sns.barplot(paid_vs_free_rating,palette='viridis')
plt.title("Average Rating: Free vs Paid Apps")
```

Conclusion: In PlayStore Paid apps has more rating then Free apps

DOES APP SIZE AFFECT INSTALLS

```python
import numpy as np
df['Size'].replace('Varies with device',np.nan,inplace=True)
```

```python
df[['Size','Installs']].corr()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='Size',y='Installs',hue='Type',palette='Set2')
plt.yscale('log')
plt.xscale('log')
plt.title("App Size vs Installs (Log Scale)")
plt.tight_layout()
```

Conclusion:App size does not strongly affect installs

```python
plt.figure(figsize=(8,5))
order= df['Category'].value_counts().sort_values(ascending=False).index
sns.countplot(df['Category'],order=order)
plt.title('Category Based Installer In Playstore')
```

```python
numerical_data=[]
for col in df.columns:
    if df[col].dtypes!='O':
        numerical_data.append(col)
numerical_data
```

Correlation analysis

```python
sns.heatmap(df[numerical_data].corr(),annot=True,cmap='coolwarm')
```

*Popular apps (more installs) tend to receive more user feedback


*Rating has very low correlation with Reviews,Installs,Price. Which means Rating is independent on installs,Price and Reviews 

Boxplot (rating by type)

```python
plot_df=df[['Type','Rating']].dropna()
plot_df.boxplot(column='Rating',by='Type')
```

Paid apps usually have a little higher ratings than free apps, but the difference is very small.

Most free and paid apps have good ratings overall. So, whether an app is free or paid does not strongly affect its rating.
